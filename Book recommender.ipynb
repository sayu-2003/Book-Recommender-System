```python
import pandas as pd

# Initialize an empty list to store valid chunks
chunks = []

# Define chunk size for reading the CSV
chunk_size = 10000

# Open the file and read line by line, handling the problematic lines
with open('data/BX-Books.csv', 'r', encoding='latin-1') as file:
    chunk = []
    line_number = 0
    problematic_lines_count = 0  # Counter for problematic lines
    max_problematic_lines = 10  # Set a maximum number of problematic lines to print
    for line in file:
        line_number += 1
        # Try to split the line; if it throws an error, print the problematic line
        try:
            data = line.strip().split(';')
            if len(data) == 10000:  # Replace YOUR_EXPECTED_NUMBER_OF_COLUMNS with the expected number of columns
                chunk.append(data)
        except Exception as e:
            # Print problematic line
            print(f"Error processing line {line_number}: {e}")
            print(f"Problematic line: {line}")
            problematic_lines_count += 1
            if problematic_lines_count >= max_problematic_lines:
                print("Too many problematic lines encountered. Stopping further output.")
                break
            continue
        
        # If chunk reaches the desired size, append it to the list and reset
        if len(chunk) >= chunk_size:
            chunks.append(pd.DataFrame(chunk))
            chunk = []

# Append the remaining chunk if it's not empty
if chunk:
    chunks.append(pd.DataFrame(chunk))

# Check if there are chunks to concatenate
if chunks:
    # Concatenate all valid chunks into a single DataFrame
    books = pd.concat(chunks, ignore_index=True)
else:
    print("No valid data found in the file.")
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```
